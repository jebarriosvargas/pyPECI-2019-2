````python
fig, axes = plt.subplots(ncols=2,nrows=1,figsize=(10,5))

x = linspace(0,2*pi,50)
y = np.cos(x)
z = np.sin(x)

axes[0].plot(x,y,'ro-',label="cos")

axes[0].legend()

axes[1].plot(x,z,'bo-',linewidth=3,markersize=5,markerfacecolor='red',
             markeredgecolor='green',label='sin')

axes[1].legend()


fig.tight_layout()
````

````python
fig, axes = plt.subplots(ncols=2,nrows=1,figsize=(10,5))

x = linspace(0,2*pi,50)
y = np.cos(x)
z = np.sin(x)

axes[0].plot(x,y,'ro-',label=r"$\cos(x)$")

axes[0].legend()

axes[1].plot(x,z,'bo-',linewidth=3,markersize=5,markerfacecolor='red',
             markeredgecolor='green',label=r"$\sin(x)$")

axes[1].legend()


fig.tight_layout()
````

**Agregar etiquetas a los ejes**

````python
fig, axes = plt.subplots(ncols=2,nrows=1,figsize=(10,5))

x = linspace(0,2*pi,50)
y = np.cos(x)
z = np.sin(x)

axes[0].set_title(r"${\rm Re}[e^{ix}]$")

axes[0].plot(x,y,'ro-',label=r"$\cos(x)$")

axes[0].legend()

axes[0].set_xlabel(r"$x$")

axes[1].set_title(r"${\rm Im}[e^{ix}]$")

axes[1].plot(x,z,'bo-',linewidth=3,markersize=5,markerfacecolor='red',
             markeredgecolor='green',label=r"$\sin(x)$")

axes[1].legend()

axes[1].set_xlabel(r"$x$")

fig.tight_layout()
````

**Editar donde se ponen las marcas en el eje x**

````python
fig, axes = plt.subplots(ncols=2,nrows=1,figsize=(10,5))

x = linspace(0,2*pi,50)
y = np.cos(x)
z = np.sin(x)

axes[0].set_title(r"${\rm Re}[e^{ix}]$")

axes[0].plot(x,y,'ro-',label=r"$\cos(x)$")

axes[0].legend()

axes[0].set_xlabel(r"$x$")

axes[0].set_xticks([0,pi/4,pi/2,3*pi/4,pi,2*pi])
axes[0].set_xticklabels([r"$0$",r"$\pi/4$",
                          r"$\pi/2$",r"$3\pi/4$",
                          r"$\pi$",r"$2\pi$"],fontsize=12)

axes[0].grid()

axes[1].set_title(r"${\rm Im}[e^{ix}]$")

axes[1].plot(x,z,'bo-',linewidth=3,markersize=5,markerfacecolor='red',
             markeredgecolor='green',label=r"$\sin(x)$")

axes[1].legend()

axes[1].set_xlabel(r"$x$")

fig.tight_layout()
````

**Guardar figura en un archivo pdf**

````python
fig, axes = plt.subplots(ncols=2,nrows=1,figsize=(10,5))

x = linspace(0,2*pi,50)
y = np.cos(x)
z = np.sin(x)

axes[0].set_title(r"${\rm Re}[e^{ix}]$")

axes[0].plot(x,y,'ro-',label=r"$\cos(x)$")

axes[0].legend()

axes[0].set_xlabel(r"$x$")

axes[0].set_xticks([0,pi/4,pi/2,3*pi/4,pi,2*pi])
axes[0].set_xticklabels([r"$0$",r"$\frac{\pi}{4}$",
                          r"$\frac{\pi}{2}$",r"$\frac{3\pi}{4}$",
                          r"$\pi$",r"$2\pi$"])
axes[0].set_xlim(0,2*pi)

axes[0].grid()

axes[1].set_title(r"${\rm Im}[e^{ix}]$")

axes[1].plot(x,z,'bo-',linewidth=3,markersize=5,markerfacecolor='red',
             markeredgecolor='green',label=r"$\sin(x)$")

axes[1].legend()

axes[1].set_xlabel(r"$x$")

fig.tight_layout()


fig.savefig("Test.pdf",bbox_inches='tight')
````

**Gráficas en 3D**

````python
from mpl_toolkits.mplot3d import Axes3D
````

````python
fig  = plt.figure()
ax   = Axes3D(fig)

X    = np.arange(-5, 5, 0.25)
Y    = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R    = np.sqrt(X**2 + Y**2)
Z    = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet)
````
